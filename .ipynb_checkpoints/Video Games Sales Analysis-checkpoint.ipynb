## Video Games Sales Analysis

In this project, I will be analysing video games sales data. After analysing this data, I hope to answer questions like 'Which is the most popular genre of games by region?', 'Which year had the most sales?', 'Which platform aquires the most market share?' etc. 

### Data Collection

The data used in this analysis is from Kaggle. https://www.kaggle.com/datasets/gregorut/videogamesales. This dataset contains a list of video games with sales greater than 100,000 copies. It was generated by a scrape of vgchartz.com. 

### Data Exploration and Transformation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading data
url = 'https://github.com/bhoomika-johnpedely/Video-Games-Sales-Analysis/blob/main/vgsales.csv?raw=True'
data = pd.read_csv(url)
```

```python
data.shape
```

```python
data.head(10)
```

##### Comment 1

The dataset is already sorted by Global_Sales. 'Wii Sports' in 2006 had the most sales globally, followed by 'Super Mario Bros.' in 1985 and 'Mario Kart Wii' in 2008. 

```python
#Checking for null
data.isnull().sum()
```

```python
# Checking for duplicates
data.duplicated().sum()
```

##### Comment 2

There were null values in the 'Year' and 'Publisher' columns. Since the data has sufficient data points, I have decided to remove the rows with null values and exclude them from this analysis. 

```python
#Dropping null
df = pd.DataFrame(data.dropna())
```

```python
#Changing column names
new_column_names = {
    'NA_Sales': 'North America',
    'EU_Sales': 'EU',
    'JP_Sales': 'Japan',
    'Other_Sales': 'Other',
    'Global_Sales': 'Global Sales'
}

# Rename the columns
df.rename(columns=new_column_names, inplace=True)
```

```python
df.info()
```

##### Comment 3

The 'Year' column has a decimal after the given year. I will be manipulating the data in the 'Year' column to remove the decimal and convert it to string. 

```python
#Transforming the data in 'Year'
df['Year'] = df['Year'].astype(str).str.split('.').str[0]
```

```python
#Selecting the columns for region
df_regions = df.iloc[:,6:10]
df_regions.head()
```

```python
# Sales by Region
colors = sns.color_palette("muted")

df_regions.sum().plot(kind='bar', color = colors)
plt.xlabel('Region')
plt.ylabel('Total Video Games Sales')
plt.title('Total Sales by Region')
plt.show()
```

##### Comment 4

North America has had the most video games sales. In the given visualization, we can see that it is almost double than that in the EU.

```python
# Sales by Year
sales_by_year = df.groupby('Year')['Global Sales'].sum()

sales_by_year.plot(kind='bar', color = colors, figsize = (10,8))
plt.xlabel('Year')
plt.ylabel('Total Video Games Sales')
plt.title('Total Sales by Year')
plt.show()
```

##### Comment 5

Video games sales peaked in 2008 and 2009 and then fell down back to where it was in the '90s. There is no data between 2017 and 2020 and these years show almost no sales. This may indicate missing data. Overall, video games sales went up in the early 2000s and started falling after 2008. 

```python
# Sales by Platform
sales_by_platform = df.groupby('Platform')['Global Sales'].sum()
sales_by_platform = sales_by_platform.sort_values(ascending = True)

# combining the smaller segments for better readability in the pie chart
threshold = sales_by_platform.sum() * 0.01
small_segments = sales_by_platform[sales_by_platform < threshold]

combined_segments = small_segments.sum()
all_segments = sales_by_platform[sales_by_platform >= threshold]
all_segments['Other'] = combined_segments
```

```python
# Pie chart showing which platforms have the highest sales
plt.figure(figsize = (9,7))
plt.pie(all_segments, labels= all_segments.index , colors=colors, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Gaming Platform Market Distribution')
plt.show()
```

##### Comment 6

This pie chart shows that according to this dataset (with data up till 2016), the most popular platform for video games is PS2 followed by X360, PS3 and Wii. 

```python
# Most popular platform by region
regions = ['North America','EU','Japan','Other']

for region in regions:
    platform_by_region = df.groupby('Platform')[region].sum()
    platform_by_region= platform_by_region.sort_values(ascending = True)
    
    share_threshold = platform_by_region.sum() * 0.05
    small_segments_byregion = platform_by_region[platform_by_region < share_threshold]

    combined_segments_byregion = small_segments_byregion.sum()
    all_segments_byregion = platform_by_region[platform_by_region >= threshold]
    all_segments_byregion['Other'] = combined_segments_byregion
    
    plt.figure(figsize = (5,3))
    plt.pie(all_segments_byregion, labels= all_segments_byregion.index , colors=colors, autopct='%1.1f%%', startangle=140)
    plt.axis('equal')
    plt.title(f'Most Popular Platform in {region}')
    plt.show()
```

##### Comment 7

The most popular platform for video games differ by region. X360 is most popular in North America, followed by PS2 and Wii. In the EU, PS3 is most popular, followed by PS2 and X360. DS is the most popular platform in Japan, followed by PS and PS2. In other areas, PS2 is the most popular gaming platform. 

```python
# Genre by Region
region = ['North America','EU','Japan','Other']

for region in regions:
    df_genre = df.groupby('Genre')[region].sum()
    sorted_df_genre = df_genre.sort_values(ascending = False)
    
    plt.figure(figsize = (5,3))
    
    sorted_df_genre.plot(kind = 'bar', color = colors)
    plt.title(f'Most Popular Genre in {region}')
    plt.ylabel('Video Games Sales')
    plt.show()
```

##### Comment 8

'Action' is the most popular video games genre followed by 'Sports' and 'Shooter' everywhere except Japan. By far, the most popular genre in Japan is 'Role-Playing' which ranks low in all other markets. 

```python
# Biggest Publishers
df.head()
df_publisher= df.groupby(['Publisher']).size().sort_values(ascending = False)
biggest_publishers= df_publisher.head(10)

# Pie Chart showing the biggest publishers
plt.figure(figsize = (9,7))
plt.pie(biggest_publishers, labels= biggest_publishers.index , colors=colors, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Biggest Publishers')
plt.show()
```

##### Comment 9

'Electronic Arts' is the biggest video games publisher, followed by 'Activision', 'Namco Bandai Games' and 'Ubisoft.
